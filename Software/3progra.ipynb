```python
import tkinter as tk
from tkinter import ttk
import cv2
from PIL import Image, ImageTk
import threading

class App:
    def _init_(self, root):
        self.root = root
        self.root.title("Aplicación con Cámara")

        self.camera = cv2.VideoCapture(0)
        self.camera_thread = threading.Thread(target=self.encender_camara)

        self.create_widgets()

    def create_widgets(self):
        self.label = tk.Label(self.root, text="¿Qué modo deseas activar?")
        self.label.pack(pady=10)

        self.boton_encender_camara = ttk.Button(self.root, text="Encender Cámara", command=self.iniciar_camara)
        self.boton_encender_camara.pack(pady=5)

        self.boton_traduccion = ttk.Button(self.root, text="Modo Traducción", command=self.modo_traduccion)
        self.boton_traduccion.pack(pady=5)

        self.boton_interactivo = ttk.Button(self.root, text="Modo Interactivo", command=self.modo_interactivo)
        self.boton_interactivo.pack(pady=5)

        self.boton_apagar = ttk.Button(self.root, text="Apagar y Salir", command=self.apagar)
        self.boton_apagar.pack(pady=5)

        self.canvas = tk.Canvas(self.root, width=640, height=480)
        self.canvas.pack(pady=10)

        self.photo = None  # Referencia al objeto PhotoImage

    def encender_camara(self):
        while True:
            ret, frame = self.camera.read()
            if not ret:
                break

            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            img = Image.fromarray(frame_rgb)
            imgtk = ImageTk.PhotoImage(image=img)

            self.photo = imgtk  # Actualizar la referencia al objeto PhotoImage

            self.canvas.create_image(0, 0, anchor=tk.NW, image=imgtk)
            self.root.update()

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

    def iniciar_camara(self):
        self.camera_thread.start()

    def modo_traduccion(self):
        print("Modo Traducción Activado")
        # Código del modo de traducción

    def modo_interactivo(self):
        print("Modo Interactivo Activado")
        # Código del modo interactivo

    def apagar(self):
        self.camera.release()
        cv2.destroyAllWindows()
        self.root.destroy()

if __name__ == "_main_":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```
